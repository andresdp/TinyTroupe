# DnD Adventure with Vision

A demonstration of TinyTroupe's **vision modality** through a tabletop RPG scenario. Three agents — a Dungeon Master and two players — experience a DnD session where the DM sees maps, NPC art, and enemy illustrations, then shares them with the players.

This notebook showcases:
- `agent.see(images=...)` — showing images to an agent
- `SHOW` action — agents sharing images with each other
- How different personas (experienced DM, veteran player, new player) react to the same visual stimuli

```python
import json
import sys
sys.path.insert(0, '..')

import tinytroupe
from tinytroupe.agent import TinyPerson
from tinytroupe.environment import TinyWorld

from IPython.display import Image, display
```

```python
greg = TinyPerson.load_specification("./agents/Greg_DM.agent.json")
talia = TinyPerson.load_specification("./agents/Talia_Player.agent.json")
jordan = TinyPerson.load_specification("./agents/Jordan_Player.agent.json")

world = TinyWorld("DnD Session", [greg, talia, jordan])
world.make_everyone_accessible()

world.broadcast_internal_goal(
    "I am at Greg's place for our weekly Dungeons & Dragons session. "
    "Greg is the Dungeon Master, Talia plays a knight named Sir Aldric, "
    "and Jordan plays an elf cleric named Elara Moonwhisper. "
    "We are about to start tonight's adventure. I should stay in character, "
    "react to what the DM describes, and interact with the other players naturally."
)
```

## Scene 1 — The World Map

The DM receives the world map for tonight's campaign. He interprets it and sets the scene for the players, sharing the map with them via the `SHOW` action.

```python
IMG_DIR = "./images/dnd"

display(Image(filename=f"{IMG_DIR}/world_map.jpeg", width=500))
```

```python
greg.see(
    images=f"{IMG_DIR}/world_map.jpeg",
    description="Here is the world map you prepared for tonight's campaign. "
                "Spread it on the table, describe the setting to Talia and Jordan, "
                "and ask them where their characters want to go."
)
world.run(3)
```

## Scene 2 — The Dungeon

The party ventures into a dungeon. The DM receives the dungeon map, two NPC allies (a knight and a cleric), and a lizard enemy. He reveals them progressively to the players as the scene unfolds.

```python
display(Image(filename=f"{IMG_DIR}/dungeon_map.jpeg", width=500))
display(Image(filename=f"{IMG_DIR}/knight_npc.jpeg", width=300))
display(Image(filename=f"{IMG_DIR}/cleric_npc.jpeg", width=300))
display(Image(filename=f"{IMG_DIR}/enemy_lizard.jpeg", width=300))
```

```python
greg.see(
    images=[
        f"{IMG_DIR}/dungeon_map.jpeg",
        f"{IMG_DIR}/knight_npc.jpeg",
        f"{IMG_DIR}/cleric_npc.jpeg",
        f"{IMG_DIR}/enemy_lizard.jpeg"
    ],
    description="The party has entered a dungeon. Here is the dungeon layout, "
                "two NPCs they encounter (a knight and a cleric who offer to help), "
                "and a lizard creature deeper in. Reveal these to Talia and Jordan "
                "as the scene unfolds — set the mood, introduce the NPCs, "
                "and build up to the encounter."
)
world.run(2)
```
